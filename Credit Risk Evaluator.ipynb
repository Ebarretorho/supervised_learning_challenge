```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df
```

```python
test_df.head(20)
```

```python
# Convert categorical data to numeric and separate target feature for training data

##get unnamed and index column out of the data as they do not represent feautures

training_data = train_df.drop(columns=['Unnamed: 0', 'index'])
#training_data.head(20)

##use get dummies to convert data into numerical
X_tr = training_data.drop(columns=['loan_status'])
X_training = pd.get_dummies(X_tr)
Y_tr = training_data["loan_status"]
y_training = LabelEncoder().fit_transform(Y_tr)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
##get unnamed and index column out of the data as they do not represent feautures

testing_data = test_df.drop(columns=['Unnamed: 0', 'index'])
#training_data.head(20)

##use get dummies to convert data into numerical
X_ts = testing_data.drop(columns=['loan_status'])
X_testing = pd.get_dummies(X_ts)
Y_ts = testing_data["loan_status"]
y_testing = LabelEncoder().fit_transform(Y_ts)
```

```python
# add missing dummy variables to testing set

##Find out what columns are missing
X_training.columns.difference(X_testing.columns)

## result ----> Index(['debt_settlement_flag_Y'], dtype='object')

X_testing['debt_settlement_flag_Y'] = 0



print(X_training.shape, y_training.shape)
print(X_testing.shape, y_testing.shape)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
logisticR_ = LogisticRegression(max_iter=8500)
logisticR_.fit(X_training, y_training)

print(f"Score on the Training Data: {logisticR_.score(X_training, y_training)}")
print(f"Score on the Test Data: {logisticR_.score(X_testing, y_testing)}")
```

```python
# Train a Random Forest Classifier model and print the model score
Rforest_ = RandomForestClassifier()
Rforest_.fit(X_training, y_training)

print(f"Score on the Training Data: {Rforest_.score(X_training, y_training)}")
print(f"Score on the Test Data: {Rforest_.score(X_testing, y_testing)}")
```

```python
# Scale the data

##Scale the Training Data

tr_scaler_ = StandardScaler().fit(X_training)
X_sc_training = tr_scaler_.transform(X_training)

##Scale the Test Data
ts_scaler_ = StandardScaler().fit(X_testing)
X_sc_testing = ts_scaler_.transform(X_testing)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
logisticR_.fit(X_sc_training, y_training)

print(f"Score on the Training Data: {logisticR_.score(X_sc_training, y_training)}")
print(f"Score on the Test Data: {logisticR_.score(X_sc_testing, y_testing)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
Rforest_.fit(X_sc_training, y_training)

print(f"Score on the Training Data: {Rforest_.score(X_sc_training, y_training)}")
print(f"Score on the Test Data: {Rforest_.score(X_sc_testing, y_testing)}")
```

## Analysis

It was my prediction that Logistic Regressions was going to perform better, as I would expect that the outcome of a loan is indeed very tied to numerical values rather; but In the results on the unscaled data RandomForest ourperformed the LogisticRegressions by 5% accuracy

When scaling the data RandomForest actually became less acurate and logisticRegression improved by 9% of accuracy.
